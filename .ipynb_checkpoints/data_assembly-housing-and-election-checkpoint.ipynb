```python
# pull tables from postgresql
pp = pd.concat(pd.read_sql(q.pp_sql,con, chunksize = 500))
apts = pd.concat(pd.read_sql(q.apt_sql,con, chunksize = 500))
res = pd.concat(pd.read_sql(q.res_sql,con, chunksize = 500))
condo = pd.concat(pd.read_sql(q.condo_sql,con, chunksize = 500))
turnout = pd.concat(pd.read_sql(q.turnout_sql, con, chunksize = 500))
lookup = pd.concat(pd.read_sql_table('lookup', con, chunksize = 500))
```

```python
lookup['luitem'] = lookup['luitem'].str.rstrip().astype(int)
lookup['lutype'] = lookup['lutype'].str.rstrip().astype(int)
lookup['ludescription'] = lookup['ludescription'].str.replace(r'[^\w\s]+', '').replace('  ','')
lookup['ludescription'] = lookup['ludescription'].str.strip()
```

```python
# filter desired columns
apts = apts[cats.keep_apts]
res = res[cats.keep_res]
condo = condo[cats.keep_condo]
```

```python
# add source table identifier
apts['apt'] = 1
res['res'] = 1
condo['condo'] = 1
```

```python
lu = {}
for x in set(lookup['lutype']):
    l = lookup[lookup['lutype'] == x]
    lu[x] = pd.Series(l.ludescription.values,index=l.luitem.astype(str)).to_dict()
lu = { k : { str(x) : y for x,y in v.items() } for k,v in lu.items() }
cc = cats.usedcols
```

```python
def rcats(df,table,cc,lu):
    for col in df.columns:
        if col in cc[table].keys():
            df[col] = df[col].replace(to_replace=lu[cc[table][col]])
```

```python
rcats(res,'res',cc,lu)
rcats(apts,'apt',cc,lu)
rcats(condo,'condo',cc,lu)
```

## Turnout Data

```python
# create mean election type targets
```

```python
et = {'pres_p' : [[2008,2012,2016],2],
'pres_g' : [[2008,2012,2016],1],
'mt_p' : [[2010,2014,2018],2],
'mt_g' : [[2010,2014,2018],1],
'odd_p' : [[2009,2011,2013,2015,2017],2],
'odd_g' : [[2009,2011,2013,2015,2017],1]}
```

```python
etd = []
for k,v in et.items():
    for y in v[0]:
        etd.append([k,y,v[1]])
```

```python
etd = pd.DataFrame(etd, columns = ['et','year','type'])
etd['type'].replace([1,2],['G','P'], inplace=True)
```

```python
turnout = turnout.merge(etd,left_on=['year','type'], right_on=['year','type'])
```

```python
sns.boxenplot(x=turnout['et'],y=turnout['turnout'])
```

```python
turnout['y'] = turnout['year'].astype(str)
```

```python
pm = pp.merge(turnout,how='outer',left_on=['name','year'],right_on=['name','y'])
```

```python
pm['pin'] = pm['major'] + pm['minor']
```

```python
pm = pm[['pin','et','turnout']]
```

```python
pm = pm.groupby(by=['pin','et'],as_index=False).mean().pivot(index='pin', columns='et', values='turnout')
```
